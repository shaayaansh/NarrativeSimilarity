```python
from pathlib import Path
import json
import pandas as pd

base = Path.cwd()
modeling_dir = base / '..' / 'data' / 'Modeling Data'
alignment_path = base / '..' / 'data' / 'Alignment' / 'sampled_story_alignments.json'

modeling_data_overall = pd.read_csv(modeling_dir / 'modeling_data_overall.csv')
modeling_data_structural = pd.read_csv(modeling_dir / 'modeling_data_structural.csv')

with open(alignment_path, 'r', encoding='utf-8') as f:
    sampled_story_alignments = json.load(f)

print('overall shape:', modeling_data_overall.shape)
print('structural shape:', modeling_data_structural.shape)
print('alignment records:', len(sampled_story_alignments))
```

```python
print('modeling_data_overall columns:')
print(modeling_data_overall.columns.tolist())
modeling_data_overall.head()
```

```python
print('modeling_data_structural columns:')
print(modeling_data_structural.columns.tolist())
modeling_data_structural.head()
```

```python
from pprint import pformat

print('alignment records:', len(sampled_story_alignments))
print('sample alignment entry (wrapped):')
sample_entry = sampled_story_alignments[0] if sampled_story_alignments else None
if sample_entry is None:
    print(None)
else:
    print(pformat(sample_entry, width=100, sort_dicts=False))
```

```python
pairs_df = pd.read_csv(base / '..' / 'data' / 'Qualtrics_Story_Pairs_sampled_clean.csv')

alignment_df = pd.DataFrame([
    {
        'PairID': item.get('PairID'),
        'num_matches': len(item.get('alignment', {}).get('matches', [])),
        'num_unmatched_a': len(item.get('alignment', {}).get('unmatched_a', [])),
        'num_unmatched_b': len(item.get('alignment', {}).get('unmatched_b', [])),
    }
    for item in sampled_story_alignments
])

pairs_df['PairID'] = pairs_df['PairID'].astype(str).str.strip()
alignment_df['PairID'] = alignment_df['PairID'].astype(str).str.strip()

merged = alignment_df.merge(
    pairs_df[['PairID', 'bucket']],
    on='PairID',
    how='left'
)

print(f"Pairs DF Length: {len(pairs_df)} | Alignment Data Length: {len(sampled_story_alignments)}")
print(f"Merged Length: {len(merged)}")
merged.head()
```

```python
print('Aligned matches per PairID (top 10):')
print(merged[['PairID', 'num_matches']].sort_values('num_matches', ascending=False).head(10).to_string(index=False))

print('\nBucket counts:')
print(merged['bucket'].value_counts(dropna=False).to_string())

print('\nnum_matches summary stats:')
print(merged['num_matches'].describe().to_string())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
sns.boxplot(data=merged, x='bucket', y='num_matches')
plt.title('Structural Matches by Bucket')
plt.ylabel('Number of Matched Event Groups')
plt.xlabel('Bucket')
plt.show()
```
